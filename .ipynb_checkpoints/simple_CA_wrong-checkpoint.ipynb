```python
class State():
    def __init__(self, state_o, rule):
        self.state = state_o
        self.all_state =[]
        self.all_state.append(self.state)
        self.rule = {'111': 0,
                    '110': 0,
                    '101': 1,
                    '100': 1,
                    '011': 0,
                    '010': 0,
                    '001': 1,
                    '000': 0,}
        self.int_to_rule(rule)
    
    # reversed?
    def int_to_rule(self, rule):
        bi = '{:08}'.format(int("{0:b}".format(rule)))
        for i, v in enumerate(reversed(str(bi))):
            self.rule['{:03}'.format(int("{0:b}".format(i)))] = int(v)
            
        
        
    def key_func(self, keystr):
        return self.rule[keystr]

        
    def to_str(self, set):
        return str(set[0])+str(set[1])+str(set[2])
    
    def pre_state(self, state):

        l = state[0::1]
        c = state[1::1]
        r = state[2::1]
        
        return zip(l, c, r)

        
    def next_state(self, state):
        
        str_list = []
        for i in list(self.pre_state(state)):
            str_list.append(self.to_str(i))

        new_state = []
        for i in str_list:
            new_state.append(self.key_func(i))

        return new_state
    
    def change_state(self, state):
        new_state = self.next_state(state)
        new_state.insert(0, 0)
        new_state.append(0)
        
        self.state = new_state
        self.all_state.append(new_state)
```

```python
import random
length =int(50)
state_o = [random.choice([1, 0]) for i in range(length)]
state_o = [i % 2 for i in range(length)]
state_o = [0 for i in range(length)]
state_o[int(length/2)] = 1

a_state = State(state_o, 18)
```
