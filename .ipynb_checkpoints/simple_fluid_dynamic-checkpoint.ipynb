```python
import sys, time
import matplotlib.pyplot as plt
from IPython.display import HTML
import numpy as np


%matplotlib inline



class SFD:
    sample_points = 100;
    sample_distance = 2 / (sample_points - 1)
    u = np.ones(sample_points)
    dt = .005

    # setting up initial condition
    wave_speed = 1
    
    @staticmethod
    def r2i(i):
        return int(i / SFD.sample_distance)

    def __init__(self):
        
        self.u[SFD.r2i(.5): SFD.r2i(1)] = 2
        
        self.history = self.u.copy()
        
    def step(self):
        un = self.u.copy()
        
            
        for i in range(1, self.sample_points):
            self.u[i] = un[i] - (un[i] * # constant c is replaced with un_i
                            (self.dt / self.sample_distance)) * (un[i] - un[i - 1])
            
        self.history = np.row_stack([self.history, self.u])
        
        
    
```

```python
sdf = SFD()
print(SFD.u)

x = np.linspace(0, 2, SFD.sample_points)
plt.plot(x, SFD.u)
```

```python
for i in range(100):
    sdf.step()
```

```python
import matplotlib.animation as animation

fig, ax = plt.subplots()
line, = ax.plot(np.linspace(0,2,sdf.sample_points), sdf.history[0])

def animate(i):
    line.set_ydata(sdf.history[i-1])
    return line,

ani = animation.FuncAnimation(fig, animate, np.arange(1, sdf.sample_points), interval=20)

HTML(ani.to_html5_video())
```
