```python
path_dct = './ThinkStats2/code/2002FemPreg.dct'
```

```python
with open(path_dct, 'r') as d:
    lines = d.readlines()
    
import re

column_info = []

for line in lines:
    
    match = re.search(r'_column\((\d+)\)', line)
    if match:
        start = int(match.group(1))
        vtype, name, fstring = line.split()[1:4]
        column_info.append([start, vtype, name, fstring])
        
        
    
```

```python
column_info.sort()
```

```python
column_info
```

```python
import pandas as pd
```

```python
pd.DataFrame(column_info, columns=['start', 'type', 'name', 'fstring'])
```
